```python
import shutil
import hashlib

# Function to calculate SHA-256 hash
def calculate_hash(filename):
    sha256 = hashlib.sha256()

    with open(filename, "rb") as file:
        while True:
            data = file.read(4096)
            if not data:
                break
            sha256.update(data)

    return sha256.hexdigest()


# Input file name
source_file = input("Enter the source file name: ")

# Destination (forensic copy)
destination_file = "forensic_copy.bin"

try:
    # Create bit-stream copy
    shutil.copyfile(source_file, destination_file)
    print("\nForensic copy created successfully.")

    # Calculate hashes
    original_hash = calculate_hash(source_file)
    copied_hash = calculate_hash(destination_file)

    print("\nOriginal File Hash:")
    print(original_hash)

    print("\nCopied File Hash:")
    print(copied_hash)

    # Verify integrity
    if original_hash == copied_hash:
        print("\nIntegrity Verified.")
        print("The forensic copy is identical to the original file.")
    else:
        print("\nIntegrity Verification Failed.")
        print("The copied file differs from the original.")

except FileNotFoundError:
    print("Error: Source file not found.")
```
