```python

from_header = input("Enter From address: ")
return_path = input("Enter Return-Path address: ")
received = input("Enter Received server: ")

print("\nEmail Header Analysis")
print("-" * 30)

print("From        :", from_header)
print("Return-Path :", return_path)
print("Received    :", received)

if from_header.lower() != return_path.lower():
    print("\nResult: Possible Email Spoofing Detected")
else:
    print("\nResult: No Spoofing Detected")
```
