```python
# Simulate Recovery of Deleted Files by Scanning Byte Patterns

filename = "sample.pdf"

try:
    with open(filename, "rb") as file:
        header = file.read(8)

    if header.startswith(b"%PDF"):
        print("PDF Document signature found at byte 0")

    elif header.startswith(b"\xFF\xD8\xFF"):
        print("JPEG Image signature found at byte 0")

    elif header.startswith(b"\x89PNG"):
        print("PNG Image signature found at byte 0")

    elif header.startswith(b"PK\x03\x04"):
        print("ZIP Archive signature found at byte 0")

    else:
        print("No recognizable file signatures found.")

except FileNotFoundError:
    print("Error: File not found.")
```
