```python
import hashlib

# Create the evidence file automatically
with open("evidence.txt", "w") as file:
    file.write("This is a sample digital evidence for chain of custody verification.")

# Function to calculate SHA-256 hash
def calculate_hash(filename):
    sha256 = hashlib.sha256()

    with open(filename, "rb") as file:
        while True:
            data = file.read(4096)
            if not data:
                break
            sha256.update(data)

    return sha256.hexdigest()

# Evidence file
filename = "evidence.txt"

# Calculate current hash
current_hash = calculate_hash(filename)

print("Evidence File Created Successfully.")
print("\nCurrent SHA-256 Hash:")
print(current_hash)

# For demonstration, use the same hash as the original
original_hash = current_hash

print("\nOriginal SHA-256 Hash:")
print(original_hash)

# Verify integrity
if current_hash == original_hash:
    print("\nEvidence Integrity Verified.")
    print("Chain of Custody is Maintained.")
else:
    print("\nEvidence Integrity Compromised!")
    print("Chain of Custody is Broken.")
```
