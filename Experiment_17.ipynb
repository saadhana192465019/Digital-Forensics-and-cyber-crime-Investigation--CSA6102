```python
import pandas as pd
import time

print("=" * 75)
print("EXPERIMENT 1.4 - HACKING ATTACK SIMULATION (ETHICAL)")
print("=" * 75)

# ============================================================
# SAMPLE USER DATABASE
# ============================================================

users = {
    "admin": "Admin@123",
    "student": "Student@2024",
    "guest": "Guest123"
}

# ============================================================
# PASSWORD BRUTE FORCE SIMULATION
# ============================================================

print("\n")
print("=" * 75)
print("1. PASSWORD BRUTE FORCE SIMULATION")
print("=" * 75)

username = input("Enter username: ")

password_list = [
    "123456",
    "password",
    "admin",
    "welcome",
    "Admin123",
    "Admin@123"
]

logs = []

for attempt,password in enumerate(password_list,1):

    print(f"Attempt {attempt} : {password}")

    time.sleep(0.5)

    if username in users and users[username] == password:

        print("Password Found")
        logs.append([username,password,"Success"])
        break

    else:

        print("Invalid Password")
        logs.append([username,password,"Failed"])

# ============================================================
# WEAK PASSWORD ANALYSIS
# ============================================================

print("\n")
print("=" * 75)
print("2. WEAK PASSWORD ANALYSIS")
print("=" * 75)

passwords = [
    "123456",
    "password",
    "qwerty",
    "welcome",
    "Admin@123",
    "abcd1234",
    "student",
    "guest"
]

weak=[]

for pwd in passwords:

    if len(pwd)<8 or pwd.isalpha() or pwd.isdigit():

        weak.append(pwd)

print("Weak Passwords Detected\n")

for i in weak:

    print(i)

# ============================================================
# SQL INJECTION DETECTION
# ============================================================

print("\n")
print("=" * 75)
print("3. SQL INJECTION DETECTION")
print("=" * 75)

payloads=[

"' OR '1'='1",

"' UNION SELECT * FROM users --",

"admin' --",

"SELECT * FROM users",

"John"

]

sql_keywords=[
" OR ",
"UNION",
"SELECT",
"--",
"DROP",
"INSERT",
"UPDATE"
]

for payload in payloads:

    detected=False

    for keyword in sql_keywords:

        if keyword.lower() in payload.lower():

            detected=True

    if detected:

        print(payload," --> SQL Injection Pattern Detected")

    else:

        print(payload," --> Safe Input")

# ============================================================
# DIRECTORY TRAVERSAL DETECTION
# ============================================================

print("\n")
print("=" * 75)
print("4. DIRECTORY TRAVERSAL DETECTION")
print("=" * 75)

paths=[

"../../etc/passwd",

"..\\..\\windows\\system32",

"/images/logo.png",

"../../../secret.txt",

"documents/report.pdf"

]

for path in paths:

    if "../" in path or "..\\" in path:

        print(path," --> Directory Traversal Detected")

    else:

        print(path," --> Safe")

# ============================================================
# ATTACK PATTERN SIGNATURES
# ============================================================

print("\n")
print("=" * 75)
print("5. ATTACK PATTERN SIGNATURES")
print("=" * 75)

signatures={

"Brute Force":"Multiple failed login attempts",

"SQL Injection":"UNION OR SELECT '--'",

"Directory Traversal":"../ or ..\\",

"Weak Password":"Length<8 or common password"

}

for attack,signature in signatures.items():

    print("\nAttack :",attack)
    print("Signature :",signature)

# ============================================================
# ATTACK LOGS
# ============================================================

print("\n")
print("=" * 75)
print("6. ATTACK PATTERN LOGS")
print("=" * 75)

attack_logs=pd.DataFrame(logs,columns=[
"Username",
"Password Tried",
"Result"
])

print(attack_logs)

# ============================================================
# VULNERABILITY REPORT
# ============================================================

print("\n")
print("=" * 75)
print("7. VULNERABILITY CLASSIFICATION REPORT")
print("=" * 75)

report=pd.DataFrame({

"Vulnerability":[
"Brute Force",
"Weak Password",
"SQL Injection",
"Directory Traversal"
],

"Risk Level":[
"High",
"Medium",
"Critical",
"High"
],

"Recommendation":[
"Enable Account Lockout",
"Use Strong Password Policy",
"Use Prepared Statements",
"Validate File Paths"
]

})

print(report)

# ============================================================
# WAF RULES
# ============================================================

print("\n")
print("=" * 75)
print("8. WAF RULE SUGGESTIONS")
print("=" * 75)

rules=[

"Block repeated failed login attempts",

"Block SQL keywords such as UNION, SELECT, DROP",

"Block '../' and '..\\\\' sequences",

"Enforce password complexity",

"Enable CAPTCHA after multiple failures"

]

for i,rule in enumerate(rules,1):

    print(i,".",rule)

# ============================================================
# SAVE REPORT
# ============================================================

attack_logs.to_csv("Attack_Pattern_Report.csv",index=False)

print("\n")
print("=" * 75)
print("CSV REPORT SAVED SUCCESSFULLY")
print("File : Attack_Pattern_Report.csv")
print("=" * 75)
```
