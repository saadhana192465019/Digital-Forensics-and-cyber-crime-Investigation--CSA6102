```python
# Phishing Email Detection

sender = input("Enter sender email: ")
subject = input("Enter email subject: ")
body = input("Enter email body: ")

score = 0

# Check suspicious sender
if ("@gmail.com" not in sender and
    "@yahoo.com" not in sender and
    "@outlook.com" not in sender and
    "@company.com" not in sender):
    score += 1

# Check suspicious subject
subject = subject.lower()
if ("urgent" in subject or
    "verify" in subject or
    "winner" in subject or
    "account" in subject or
    "password" in subject):
    score += 1

# Check suspicious body
body = body.lower()
keywords = ["click", "login", "verify", "bank", "password", "otp", "credit card"]

for word in keywords:
    if word in body:
        score += 1

# Final Decision
if score >= 2:
    print("\nResult: Phishing Email Detected")
else:
    print("\nResult: Legitimate Email")
```
