```python
# Program to identify file type using magic numbers

magic_numbers = {
    b'\xFF\xD8\xFF': 'JPEG Image',
    b'\x89PNG': 'PNG Image',
    b'GIF87a': 'GIF Image',
    b'GIF89a': 'GIF Image',
    b'%PDF': 'PDF Document',
    b'PK\x03\x04': 'ZIP Archive',
    b'Rar!': 'RAR Archive'
}

filename = input("Enter the file name: ")

try:
    with open(filename, "rb") as file:
        header = file.read(8)

    found = False

    for magic, filetype in magic_numbers.items():
        if header.startswith(magic):
            print("\nFile Type:", filetype)
            found = True
            break

    if not found:
        print("\nUnknown or Unsupported File Type")

except FileNotFoundError:
    print("Error: File not found.")
```
