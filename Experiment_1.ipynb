```python
import os
import time
from datetime import datetime

# File name
filename = "metadata_demo.txt"

# Create the file if it does not exist
with open(filename, "w") as file:
    file.write("Initial content.\n")

    print("=== BEFORE MODIFICATION ===")
    print("Created Time     :", datetime.fromtimestamp(os.path.getctime(filename)))
    print("Modified Time    :", datetime.fromtimestamp(os.path.getmtime(filename)))
    print("Last Access Time :", datetime.fromtimestamp(os.path.getatime(filename)))

    # Wait for a few seconds
    time.sleep(5)

    # Modify the file
    with open(filename, "a") as file:
        file.write("This line was added later.\n")

        print("\nFile has been modified...\n")

        print("=== AFTER MODIFICATION ===")
        print("Created Time     :", datetime.fromtimestamp(os.path.getctime(filename)))
        print("Modified Time    :", datetime.fromtimestamp(os.path.getmtime(filename)))
        print("Last Access Time :", datetime.fromtimestamp(os.path.getatime(filename)))

        print("\nObservation:")
        print("The modification timestamp has changed after editing the file.")
        print("This metadata acts as digital evidence, demonstrating Locard's Exchange Principle.")
```
