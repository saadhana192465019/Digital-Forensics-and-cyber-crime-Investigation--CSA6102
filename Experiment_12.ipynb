```python
# Program to Extract Image Metadata (Digital Forensics)
# Install Pillow once using:
# pip install Pillow

from PIL import Image
from PIL.ExifTags import TAGS

# Function to create a sample image
def create_sample_image(path):
    # Creates a small test image
    img = Image.new("RGB", (100, 100), color="blue")
    img.save(path)

# Function to display image metadata
def show_metadata(path):
    img = Image.open(path)

    print("File:", path)
    print("Format:", img.format)
    print("Size:", img.size)
    print("Mode:", img.mode)

    # Read EXIF metadata
    exif_data = img.getexif()

    if exif_data:
        print("\nEXIF Metadata:")
        for tag_id, value in exif_data.items():
            tag = TAGS.get(tag_id, tag_id)
            print(f"{tag}: {value}")
    else:
        print("\nNo EXIF metadata found (this sample image has none).")
        print("Real photos from phones/cameras typically contain GPS,")
        print("device model, and timestamp data - valuable forensic evidence.")

# Main Program
create_sample_image("sample.jpg")
show_metadata("sample.jpg")
```
