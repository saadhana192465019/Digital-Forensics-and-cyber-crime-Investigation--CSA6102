```python
import os
import csv
from collections import defaultdict

# -------------------------------
# Create a simulated network log
# -------------------------------
filename = "network_log.csv"

if not os.path.exists(filename):
    with open(filename, "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Source_IP", "Destination_IP", "Port"])

        # Normal traffic
        writer.writerow(["192.168.1.10", "192.168.1.100", 80])
        writer.writerow(["192.168.1.10", "192.168.1.100", 443])
        writer.writerow(["192.168.1.20", "192.168.1.101", 22])
        writer.writerow(["192.168.1.20", "192.168.1.101", 80])

        # Simulated port scan from one attacker
        for port in range(20, 36):   # 16 different ports
            writer.writerow(["192.168.1.50", "192.168.1.100", port])

print("Simulated network log created:", filename)

# ------------------------------------
# Detect Port Scanning
# ------------------------------------
scan_threshold = 10

ports_accessed = defaultdict(set)

with open(filename, "r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        src = row["Source_IP"]
        port = int(row["Port"])
        ports_accessed[src].add(port)

print("\n------ Port Scan Detection Report ------")

found = False

for ip, ports in ports_accessed.items():
    if len(ports) >= scan_threshold:
        print(f"ALERT: Possible Port Scan Detected!")
        print(f"Source IP        : {ip}")
        print(f"Ports Accessed   : {len(ports)}")
        print(f"Port Numbers     : {sorted(ports)}")
        print("----------------------------------------")
        found = True

if not found:
    print("No port scanning activity detected.")
```
