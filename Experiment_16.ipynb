```python
import pandas as pd

# =====================================================
# CREATE SAMPLE RANSOMWARE DATABASE
# =====================================================

ransomware_data = [

    {
        "Family":"WannaCry",
        "Extension":".WNCRY",
        "Domain":"iuqerfsodp9ifjaposdfjhgosurijfaewrwergwea.com",
        "IP":"192.168.10.15",
        "RansomNote":"Please Read Me.txt",
        "Encryption":"AES-128",
        "Bitcoin":"1A2B3C4D"
    },

    {
        "Family":"LockBit",
        "Extension":".lockbit",
        "Domain":"lockbit-support.onion",
        "IP":"185.220.101.5",
        "RansomNote":"Restore-My-Files.txt",
        "Encryption":"AES-256",
        "Bitcoin":"3F4G5H6J"
    },

    {
        "Family":"Conti",
        "Extension":".conti",
        "Domain":"contirecovery.onion",
        "IP":"91.219.236.15",
        "RansomNote":"CONTI_README.txt",
        "Encryption":"RSA-2048",
        "Bitcoin":"7H8J9K0L"
    },

    {
        "Family":"Ryuk",
        "Extension":".ryk",
        "Domain":"ryukhelp.onion",
        "IP":"104.18.30.50",
        "RansomNote":"RyukReadMe.txt",
        "Encryption":"RSA+AES",
        "Bitcoin":"5D6E7F8G"
    },

    {
        "Family":"Maze",
        "Extension":".maze",
        "Domain":"mazehelp.onion",
        "IP":"172.16.20.10",
        "RansomNote":"DECRYPT-FILES.txt",
        "Encryption":"AES-256",
        "Bitcoin":"9P8Q7R6S"
    }

]

df = pd.DataFrame(ransomware_data)

print("="*70)
print("RANSOMWARE FAMILY PROFILE DATABASE")
print("="*70)

print(df)

print("\n")

# =====================================================
# DISPLAY FAMILY PROFILES
# =====================================================

print("="*70)
print("RANSOMWARE FAMILY DETAILS")
print("="*70)

for index,row in df.iterrows():

    print("\nFamily :",row["Family"])
    print("Extension :",row["Extension"])
    print("Encryption :",row["Encryption"])
    print("Domain :",row["Domain"])
    print("IP Address :",row["IP"])
    print("Ransom Note :",row["RansomNote"])
    print("Bitcoin Wallet :",row["Bitcoin"])
    print("-"*70)

# =====================================================
# IoC COLLECTION
# =====================================================

print("\n")
print("="*70)
print("INDICATORS OF COMPROMISE (IoCs)")
print("="*70)

ioc_database = {}

for index,row in df.iterrows():

    ioc_database[row["Family"]] = {

        "Extension":row["Extension"],
        "Domain":row["Domain"],
        "IP":row["IP"],
        "Bitcoin":row["Bitcoin"],
        "Ransom Note":row["RansomNote"]

    }

for family,details in ioc_database.items():

    print("\n",family)

    for key,value in details.items():

        print("   ",key,":",value)

# =====================================================
# RANSOM NOTE ANALYSIS
# =====================================================

print("\n")
print("="*70)
print("RANSOM NOTE CHARACTERISTIC ANALYSIS")
print("="*70)

keywords = [

"decrypt",
"bitcoin",
"payment",
"recover",
"files",
"restore",
"key",
"contact"

]

notes = [

"Your files are encrypted. Pay Bitcoin to recover your files.",

"Restore your files after payment.",

"Decrypt documents using the private key.",

"Contact us immediately to recover all files."

]

for note in notes:

    print("\nRansom Note:")
    print(note)

    found=[]

    for word in keywords:

        if word.lower() in note.lower():

            found.append(word)

    print("Detected Keywords :",found)

# =====================================================
# YARA-LIKE RULES
# =====================================================

print("\n")
print("="*70)
print("YARA-LIKE DETECTION RULES")
print("="*70)

for index,row in df.iterrows():

    print(f"""

rule {row["Family"]}_Detection
{{
    meta:

        author = "Student"

        description = "Detects {row["Family"]} ransomware"

    strings:

        $extension = "{row["Extension"]}"

        $domain = "{row["Domain"]}"

        $note = "{row["RansomNote"]}"

    condition:

        any of them
}}

""")

# =====================================================
# SEARCH FOR IoCs
# =====================================================

print("\n")
print("="*70)
print("SEARCH RANSOMWARE IoCs")
print("="*70)

search = input("Enter Domain/IP/Extension : ")

found=False

for index,row in df.iterrows():

    if search.lower() in str(row.values).lower():

        found=True

        print("\nMATCH FOUND")
        print(row)

if not found:

    print("\nNo matching IoC found.")

# =====================================================
# SIMULATED RANSOMWARE ATTACK TIMELINE
# =====================================================

print("\n")
print("="*70)
print("SIMULATED RANSOMWARE DEPLOYMENT TIMELINE")
print("="*70)

timeline=[

("08:00","Phishing Email Delivered"),

("08:03","Victim Opens Email"),

("08:04","Malicious Attachment Executed"),

("08:05","Command and Control Communication"),

("08:07","Privilege Escalation"),

("08:09","System Enumeration"),

("08:12","File Encryption Started"),

("08:20","Files Renamed with New Extension"),

("08:22","Ransom Note Created"),

("08:25","Desktop Wallpaper Changed"),

("08:27","Encryption Completed")

]

for time,event in timeline:

    print(time," ---> ",event)

# =====================================================
# SAVE DATABASE
# =====================================================

df.to_csv("Ransomware_IoC_Database.csv",index=False)

print("\n")
print("="*70)
print("DATABASE SAVED SUCCESSFULLY")
print("File Name : Ransomware_IoC_Database.csv")
print("="*70)
```
