```python
# Brute-Force Intrusion Detection

failed_attempts = {}

n = int(input("Enter the number of login records: "))

for i in range(n):
    username = input("Enter username: ")
    status = input("Enter login status (Success/Failed): ")

    if status.lower() == "failed":
        if username in failed_attempts:
            failed_attempts[username] += 1
        else:
            failed_attempts[username] = 1

print("\nBrute-Force Detection Report")
print("-" * 35)

attack_found = False

for user, count in failed_attempts.items():
    print(user, ":", count, "failed login attempts")

    if count >= 3:
        print(">> ALERT: Possible Brute-Force Attack Detected for", user)
        attack_found = True

if not attack_found:
    print("\nNo Brute-Force Attack Detected.")
```
