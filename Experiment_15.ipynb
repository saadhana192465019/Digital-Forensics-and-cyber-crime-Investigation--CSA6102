```python
import pandas as pd
import hashlib
from collections import Counter

# ==========================
# Load Dataset
# ==========================

df = pd.read_csv("sample_files.csv")

# ==========================
# Generate Hashes
# ==========================

def md5_hash(text):
    return hashlib.md5(text.encode()).hexdigest()

def sha1_hash(text):
    return hashlib.sha1(text.encode()).hexdigest()

def sha256_hash(text):
    return hashlib.sha256(text.encode()).hexdigest()

df["MD5"] = df["Content"].apply(md5_hash)
df["SHA1"] = df["Content"].apply(sha1_hash)
df["SHA256"] = df["Content"].apply(sha256_hash)

# ==========================
# Sample Malware Database
# ==========================

known_malware = [
    "bank_update.exe",
    "trojan.exe",
    "ransomware.exe",
    "keylogger.exe"
]

df["Known_Malware"] = df["File_Name"].isin(known_malware)

# ==========================
# Hash-Based Detection
# ==========================

malware_hashes = set(df[df["Known_Malware"]]["SHA256"])

def detect(hash_value):
    if hash_value in malware_hashes:
        return "Malware"
    return "Safe"

df["Detection"] = df["SHA256"].apply(detect)

# ==========================
# Hash Collision Analysis
# ==========================

md5_collision = df["MD5"].duplicated().sum()
sha1_collision = df["SHA1"].duplicated().sum()
sha256_collision = df["SHA256"].duplicated().sum()

# ==========================
# Hash Family Clustering
# ==========================

def family(file):
    if "trojan" in file.lower():
        return "Trojan"
    elif "ransom" in file.lower():
        return "Ransomware"
    elif "keylogger" in file.lower():
        return "Keylogger"
    elif "bank" in file.lower():
        return "Banking Malware"
    else:
        return "Benign"

df["Family"] = df["File_Name"].apply(family)

# ==========================
# Detection Statistics
# ==========================

total = len(df)

malware = (df["Detection"]=="Malware").sum()

safe = (df["Detection"]=="Safe").sum()

rate = malware/total*100

# ==========================
# Print Results
# ==========================

print("\n==============================")
print("MALWARE HASH DATABASE")
print("==============================\n")

print(df[[
    "File_Name",
    "MD5",
    "SHA1",
    "SHA256",
    "Detection",
    "Family"
]])

print("\n==============================")
print("DETECTION STATISTICS")
print("==============================")

print("Total Files:", total)
print("Malware:", malware)
print("Safe:", safe)
print("Detection Rate:", round(rate,2),"%")

print("\n==============================")
print("HASH COLLISIONS")
print("==============================")

print("MD5 Collisions:", md5_collision)
print("SHA1 Collisions:", sha1_collision)
print("SHA256 Collisions:", sha256_collision)

print("\n==============================")
print("HASH FAMILY CLUSTERING")
print("==============================")

print(df["Family"].value_counts())

# ==========================
# Save Malware Database
# ==========================

df.to_csv("Malware_Hash_Database.csv",index=False)

print("\nMalware database saved as Malware_Hash_Database.csv")
```
